```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime 
```

```python
# 1
df_data = pd.read_excel('airlines.xlsx', parse_dates=['Date_of_Journey'])
df_data
```

```python
# 2

df_data = df_data.dropna()
```

```python
# 3

df_data.isnull().values.any()
```

```python
# 4

df_data['Date_of_Journey'] = pd.to_datetime(df_data['Date_of_Journey'])
df_data['Dep_Time'] = pd.to_datetime(df_data['Dep_Time'])
df_data['Arrival_Time'] = pd.to_datetime(df_data['Arrival_Time'])

df_data.dtypes
```

```python
# 5

## date

jDay = df_data['Date_of_Journey'].dt.day
jMonth = df_data['Date_of_Journey'].dt.month
jYear = df_data['Date_of_Journey'].dt.year

df_data['Journey_day'] = jDay
df_data['Journey_month'] = jMonth
df_data['Journey_year'] = jYear

## time

dHour = df_data['Dep_Time'].dt.hour
dMinute = df_data['Dep_Time'].dt.minute

aHour = df_data['Arrival_Time'].dt.hour
aMinute = df_data['Arrival_Time'].dt.minute

df_data['Dep_Time_hour'] = dHour
df_data['Dep_Time_minute'] = dMinute
df_data['Arrival_Time_hour'] = aHour
df_data['Arrival_Time_minute'] = aMinute

df_data.dtypes
```

```python
# 6

def giveDescription(time):

    if (time >= 4 and time < 8):
        return 'Early Morning'

    elif (time >= 8 and time < 12):
        return 'Morning'

    if (time >= 12 and time < 16):
        return 'Noon'

    if (time >= 16 and time < 18):
        return 'Evening'

    if (time >= 18 and time < 24):
        return 'Night'

    elif (time >= 0 and time < 4):
        return 'Late Night'

desc = df_data['Dep_Time_hour'].apply(giveDescription)
df_data['dep_description'] = desc
df_data
```

```python
# 8

xAxis = df_data['dep_description'].drop_duplicates()
yAxis = df_data['dep_description'].value_counts()

plt.bar(xAxis, yAxis)
```

```python
# 9
dHour = df_data['Duration'].str.extract('(\d+)h', expand=False).astype(float) * 60
dMin = df_data['Duration'].str.extract('(\d+)m', expand=False).astype(float)

df_data['Duration_Total_mins'] = dHour.add(dMin, fill_value=0).astype(int)
df_data['Duration_hours'] = dHour/60
df_data['Duration_mins'] = dMin

df_data
```

```python
# 10

xAxis = df_data['Duration_Total_mins']
yAxis = df_data['Price']

plt.scatter(xAxis, yAxis)
```

```python
# 11

xAxis = df_data['Duration_Total_mins']
yAxis = df_data['Price']

m, b = np.polyfit(xAxis, yAxis, 1)

plt.scatter(xAxis, yAxis, s=2)

x_fit = np.linspace(xAxis.min(), xAxis.max(), 100)
y_fit = m * x_fit + b
plt.plot(x_fit, y_fit, 'r--', label="best fit line")
plt.legend()
```

```python
# 12

xAxis = df_data['Duration_Total_mins']
yAxis = df_data['Price']

df_data['Total_Stops'].drop_duplicates()

def get_color(stop):
    c = 0
    if (stop == "non-stop"):
        c = 0

    elif (stop == "1 stop"):
        c = 1

    elif (stop == "2 stops"):
        c = 2

    elif (stop == "3 stops"):
        c = 3

    else:
        c = 4

    return c

color = df_data['Total_Stops'].apply(get_color)

plt.scatter(xAxis, yAxis, c=color, s=1)
```

```python
# 13

jet_df = df_data.loc[df_data["Airline"] == "Jet Airways"]
jet_data = jet_df.groupby(['Airline'])['Route'].value_counts().head(10)

jet_data.plot(kind="bar", figsize = [8, 6])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from patsy import dmatrices
```

```python
# 14

yAxis, xAxis = dmatrices("Price ~ Duration_Total_mins", df_data, return_type='dataframe')

x_train, x_test, y_train, y_test = train_test_split(xAxis, yAxis, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

r2 = r2_score(y_test, y_pred)
print(f'R² score: {r2}')

plt.scatter(df_data["Duration_Total_mins"], df_data["Price"], s=5, alpha=0.2)

slope = model.coef_[0][1]
intercept = model.intercept_[0]
x_fit = np.linspace(df_data["Duration_Total_mins"].min(), df_data["Duration_Total_mins"].max(), 100)
y_fit = slope * x_fit + intercept
plt.plot(x_fit, y_fit, label="Linear regression", color="red")
```

```python
# 15
df_data['Journey_weekday'] = df_data['Date_of_Journey'].dt.weekday

dimensional_df = dmatrices("Price ~ Airline * Source * Destination * Total_Stops * dep_description * Journey_month * Journey_weekday", df_data, return_type='dataframe')

x_train, x_test, y_train, y_test = train_test_split(dimensional_df[0], dimensional_df[1], test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print(model.score(x_train, y_train))
r2 = r2_score(y_test, y_pred)
print(f'R² score: {r2}')
```
